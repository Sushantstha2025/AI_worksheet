```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/drive/MyDrive/datasets/Copy of diabetes_.csv")

def train_test_split_scratch(x, y, test_size=0.3, random_seed = 42):
  np.random.seed(random_seed)
  indices = np.arange(x.shape[0])
  np.random.shuffle(indices)
  test_split_size = int(len(x) * test_size)
  test_indices = indices[ : test_split_size]
  train_indices = indices[test_split_size : ]
  x_train, x_test = x[train_indices], x[test_indices]
  y_train, y_test = y[train_indices], y[test_indices]

  return x_train, x_test, y_train, y_test

def euclidean_distance(x, y):
  if x.shape != y.shape:
    raise ValueError("Points must have the same dimensions to calculate Euclidean distance")

  distance = np.sqrt(np.sum((x-y)**2))
  return distance

def knn_predict_single(query, x_train, y_train, k=3):
  distances = [euclidean_distance(query, x) for x in x_train]
  sorted_indices = np.argsort(distances)
  nearest_indices = sorted_indices[ :k]
  nearest_labels = y_train[nearest_indices]

  prediction = np.bincount(nearest_labels).argmax()

  return prediction


def knn_predict(x_test, x_train, y_train, k=3):
  predictions = [knn_predict_single(x, x_train, y_train, k) for x in x_test]
  return np.array(predictions)

def compute_accuracy(y_true, y_pred):
  correct_predictions = np.sum(y_true == y_pred)
  total_predictions = len(y_true)
  accuracy = (correct_predictions / total_predictions) * 100
  return accuracy

x = df.drop(columns=["Outcome"]).values
y = df["Outcome"].values

# Train-test split
x_train, x_test, y_train, y_test = train_test_split_scratch(x, y)

# KNN prediction
k = 3
y_pred = knn_predict(x_test, x_train, y_train, k)

# Accuracy
accuracy = compute_accuracy(y_test, y_pred)
print(f"Accuracy of the KNN model without scaling: {accuracy:.2f}%")

#scaling
def standardize_features(X):
  mean = np.mean(X, axis = 0)
  std = np.std(X, axis = 0)
  return (X-mean)/std

X_scaled = standardize_features(x)

#train test split on scaled data
x_train_s, x_test_s, y_train_s, y_test_s = train_test_split_scratch(X_scaled, y)

y_pred_scaled = knn_predict(x_test_s, x_train_s, y_train_s, k = 3)
accuracy_scaled = compute_accuracy(y_test_s, y_pred_scaled)

print(f"Accuracy after feature scaling: {accuracy_scaled:.2f}%")

# After scaling the feature matrix using standardization, the KNN classifier showed improved performance.
# Feature scaling ensures that all variables contribute equally to distance computation, which is crucial for distance-based algorithms like KNN.

# Comparative Analysis: Original vs. Scaled Dataset

# The KNN classifier was first trained and evaluated using the original (unscaled) dataset. The model achieved an accuracy of XX%. After applying feature scaling using standardization, the KNN model was retrained and evaluated on the scaled dataset, resulting in an improved accuracy of YY%.

# Impact of Scaling on KNN Performance

# Feature scaling significantly improved the performance of the KNN classifier. Since KNN is a distance-based algorithm that relies on Euclidean distance, unscaled features with larger numeric ranges dominated the distance calculations, reducing the influence of smaller-scale features.

# Reason for Observed Accuracy Changes

# By scaling the features, all variables contributed equally to the distance computation. This led to more meaningful neighbor selection and improved classification accuracy. The improved performance demonstrates the importance of feature scaling when using distance-based machine learning algorithms such as KNN.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time


df = pd.read_csv("/content/drive/MyDrive/datasets/Copy of diabetes_.csv")

def train_test_split_scratch(x, y, test_size=0.3, random_seed = 42):
  np.random.seed(random_seed)
  indices = np.arange(x.shape[0])
  np.random.shuffle(indices)
  test_split_size = int(len(x) * test_size)
  test_indices = indices[ : test_split_size]
  train_indices = indices[test_split_size : ]
  x_train, x_test = x[train_indices], x[test_indices]
  y_train, y_test = y[train_indices], y[test_indices]

  return x_train, x_test, y_train, y_test

def euclidean_distance(x, y):
  if x.shape != y.shape:
    raise ValueError("Points must have the same dimensions to calculate Euclidean distance")

  distance = np.sqrt(np.sum((x-y)**2))
  return distance

def knn_predict_single(query, x_train, y_train, k=3):
  distances = [euclidean_distance(query, x) for x in x_train]
  sorted_indices = np.argsort(distances)
  nearest_indices = sorted_indices[ :k]
  nearest_labels = y_train[nearest_indices]

  prediction = np.bincount(nearest_labels).argmax()

  return prediction


def knn_predict(x_test, x_train, y_train, k=3):
  predictions = [knn_predict_single(x, x_train, y_train, k) for x in x_test]
  return np.array(predictions)

def compute_accuracy(y_true, y_pred):
  correct_predictions = np.sum(y_true == y_pred)
  total_predictions = len(y_true)
  accuracy = (correct_predictions / total_predictions) * 100
  return accuracy

def experiment_knn_k_values(x_train, y_train, x_test, y_test, k_values):
  accuracies = []
  times = []

  for k in k_values:
    start_time = time.time()
    predictions = knn_predict(x_test, x_train, y_train, k = k)
    end_time = time.time()

    accuracy = compute_accuracy(y_test, predictions)
    time_taken = end_time - start_time

    accuracies.append(accuracy)
    times.append(time_taken)
    print(f"k = {k} \n Accuracy = {accuracy:.2f}% \n Time = {time_taken:.4f}s \n")

  return accuracies, times



x = df.drop(columns=["Outcome"]).values
y = df["Outcome"].values

k_values = list(range(1, 16))

# Train-test split
x_train, x_test, y_train, y_test = train_test_split_scratch(x, y)
acc_orig, time_orig = experiment_knn_k_values(x_train, y_train, x_test, y_test, k_values)
acc_scaled, time_scaled = experiment_knn_k_values(x_train_s, y_train_s, x_test_s, y_test_s, k_values)


# KNN prediction
k = 3
y_pred = knn_predict(x_test, x_train, y_train, k)

# Accuracy plot
plt.plot(k_values, acc_orig, label='Original')
plt.plot(k_values, acc_scaled, label='Scaled')
plt.xlabel('k')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.title('k vs Accuracy')
plt.show()

# Time plot
plt.plot(k_values, time_orig, label='Original')
plt.plot(k_values, time_scaled, label='Scaled')
plt.xlabel('k')
plt.ylabel('Time (seconds)')
plt.legend()
plt.title('k vs Time Taken')
plt.show()


# Accuracy
accuracy = compute_accuracy(y_test, y_pred)
print(f"Accuracy of the KNN model without scaling: {accuracy:.2f}%")
```
