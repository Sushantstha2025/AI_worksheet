```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/datasets/Copy of Titanic-Dataset.csv")

df.interpolate(method="linear", axis=0, inplace=True)   # interpolated because fare and age contained null values.
df["fare_per_year"] = df["Fare"]/df["Age"]
print(df["fare_per_year"])

high_fare_age = df[df["fare_per_year"]>5]

high_fare_age_srt = high_fare_age.sort_values(by="fare_per_year", ascending=False)

result = df[["Name", "fare_per_year"]]
print(result)
```

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/datasets/Copy of Titanic-Dataset.csv")

total_fare_paid = df["Fare"].sum()

first_class_total_fare = df[df["Pclass"] == 1]["Fare"].sum()
print(first_class_total_fare)

second_class_total_fare = df[df["Pclass"] == 2]["Fare"].sum()
print(second_class_total_fare)

third_class_total_fare = df[df["Pclass"] == 3]["Fare"].sum()
print(third_class_total_fare)

fare_list = [first_class_total_fare, second_class_total_fare, third_class_total_fare]
fare_list = [fare_revenue/total_fare_paid for fare_revenue in fare_list]
print(fare_list)
```

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/datasets/Copy of Titanic-Dataset.csv")

df["age_group"] = pd.cut(df["Age"], bins=[0, 18, 65, float("inf")], labels=["child", "adult", "senior"], right=False)
print(df["age_group"])

total_passengers = len(df["Name"])
print(total_passengers)

# print(df.tail(1)["PassengerId"])  # it is much faster than calculating the length

# total_passenger_by_age = df.groupby("age_group")["PassengerId"].count()  # pandas groups each age_group and converts it into rows and assigns it an id, so simply counting the passenger id for each label gives the total number.
# print(total_passenger_by_age)

total_passenger_by_age = df["age_group"].value_counts()

total_passenger_by_age = total_passenger_by_age / total_passengers

for value in total_passenger_by_age:
  percentage_str = f"{value*100:.2f}%"
  print(percentage_str)
```
